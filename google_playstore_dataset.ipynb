```python
df_copy=df.copy()
```

```python
df_copy=df_copy.drop(df_copy.index[10472])
```

```python
df_copy[-df_copy['Reviews'].str.isnumeric()]
```

```python
#convert Review Datatype to int
df_copy['Reviews']=df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
#19000000=19000k
```

```python
df_copy['Size'].isnull().sum()
```

```python
df_copy['Size']=df_copy['Size'].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size']=df_copy['Size'].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Size']
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
char_to_remove=['+','-','$']
cols_to_clean=['Installs','Price']
for item in char_to_remove:
  for cols in cols_to_clean:
    df_copy[cols]=df_copy[cols].str.replace(',','')
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype('int')
df_copy['Price']=df_copy['Price'].astype('float')
```

```python
df_copy.info()
```

```python
# handling last update features
df_copy['Last Updated'].unique()
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
```

```python
df_copy.info()
```

```python
df_copy.to_csv('data/google_cleaned.csv')
```

```python
df_copy=df_copy.drop_duplicates(subset=['App'],keep='first')
```

```python
## EDA
```

```python
numeric_features=[features for features in df_copy.columns if df_copy[features].dtype != 'O']
categorical_features=[features for features in df_copy.columns if df_copy[features].dtype == 'O']
#print columns
print('we have {} numerical features : {}'.format(len(numeric_features),numeric_features))
print('\nwe have {} categorical features : {}'.format(len(categorical_features),categorical_features))
```

```python
##what is proportion of count data on categorical columns
for col in categorical_features:
  print(df[col].value_counts(normalize=True)*100)
  print('------------------------')
```

```python
#proportion of count data on numerical columns
plt.figure(figsize=(15,15))
plt.suptitle("Univariant Analysis of Numeric Features",fontsize=20,fontweight='bold',alpha=.8,y=1.)

for i in range(0, len(numeric_features)):
  plt.subplot(5,3,i+1)
  sns.kdeplot(x=df_copy[numeric_features[i]],shade=True,color='r')
  plt.xlabel(numeric_features[i])
  plt.tight_layout()
```

```python
#observation
'''Rating and year left skewed while Reviews,size,installs and price are right skewed'''
```

```python
#categorical Features
plt.figure(figsize=(20,15))
plt.suptitle("Univariant Analysis of Categorical Features",fontsize=20,fontweight='bold',alpha=.8,y=1.)
category=['Type','Content Rating']
for i in range(0, len(category)):
  plt.subplot(2,2,i+1)
  sns.countplot(x=df[category[i]],palette='Set2')
  plt.xlabel(category[i])
  plt.xticks(rotation=45)
  plt.tight_layout()
```
