```python
import matplotlib.pyplot as mp
```

```python
import numpy as np
```

```python
x=np.random.rand(50)
```

```python
y=np.random.rand(50)
```

```python
x
```

```python
y
```

```python
mp.scatter(x,y)
```

```python
mp.scatter(x,y ,c='#2238DC',alpha=.9,marker='X')
mp.xlabel("x axis")
mp.ylabel("y axis")
mp.title("x and y graph")
mp.grid()
```

```python
mp.figure(figsize=(4,4))
mp.scatter(x,y ,c='#2238DC',alpha=.9,marker='X')
mp.xlabel("x axis")
mp.ylabel("y axis")
mp.title("x and y graph")
mp.grid()
```

```python
x=np.linspace(1,10,100)
y=np.sin(x)
```

```python
x=['a','b','c','d','e']
y=np.random.rand(5)
```

```python
x
```

```python
y
```

```python
mp.bar(x,y,color='green')
mp.xlabel("X AXIS")
mp.title("X v/s Y BARPLOT")
mp.ylabel("Y AXIS")
mp.figure(figsize=(4,4))
```

```python

mp.barh(x,y,color='black')
mp.xlabel("X AXIS")
mp.title("X v/s Y BARPLOT")
mp.ylabel("Y AXIS")
mp.figure(figsize=(4,4))
```
