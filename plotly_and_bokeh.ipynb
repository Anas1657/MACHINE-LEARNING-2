```python
import plotly.graph_objects as go
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5],y=[3,4,5,6,7],mode='lines'))
```

```python
import seaborn as sns
tips=sns.load_dataset('tips')
fig=go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill,y=tips.tip,mode='markers'))
```

```python
tips
```

```python
fig=go.Figure()
fig.add_trace(go.Bar(x=[1,2,3,4,5],y=[3,4,5,6,7]))
```

```python
fig=go.Figure()
fig.add_trace(go.Histogram(x=tips['total_bill']))
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill,y=tips.tip,mode='markers',marker_size=10*tips['size']))
```

```python
tips
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter3d(x=tips.total_bill,y=tips.tip,z=10*tips['size']))
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter3d(x=[1,2,3,4],y=[5,6,7,8],z=[7,8,9,0],mode='lines'))
```

**BOKEH**

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure,output_file,show
from bokeh.sampledata.iris import flowers
```

```python
flowers
```

```python
output_file("test.html")
p=figure(title='test flowers')
p.xaxis.axis_label="petal_length"
p.yaxis.axis_label="petal_width"
p.circle(flowers["petal_length"],flowers["petal_width"])
show(p)
```

```python
output_file("test.html")
p=figure(title='test flowers')
p.xaxis.axis_label="petal_length"
p.yaxis.axis_label="petal_width"
p.line(flowers["petal_length"],flowers["petal_width"])
show(p)
```

```python
x=[3,4,5,6,7]
y=[4,5,6,7,4]
output_file("line.html")
p=figure(title="line graph")
p.line(x,y)
show(p)
```

```python
x=[3,4,5,6,7]
y=[4,5,6,7,4]
output_file("line.html")
p=figure(title="line graph")
p.scatter(x,y,size=20,color='red',legend_label="this is my data point")
show(p)
```
